```python
import pandas as pd
import numpy as np
from pydataset import data
```

```python
np.random.seed(123)
students = ['Sally', 'Jane', 'Suzie', 'Billy', 'Ada', 'John', 'Thomas',
            'Marie', 'Albert', 'Richard', 'Isaac', 'Alan']
math_grades = np.random.randint(low=60, high=100, size=len(students))
english_grades = np.random.randint(low=60, high=100, size=len(students))
reading_grades = np.random.randint(low=60, high=100, size=len(students))
```

```python
df = pd.DataFrame({'name': students,
                   'math': math_grades,
                   'english': english_grades,
                   'reading': reading_grades})

type(df)
```

```python
df['passing english'] = df.english >= 70
```

```python
df.sort_values('passing english')
```

```python
## duplicates are put in order of index number
```

```python
df.sort_values(['passing english', 'name'])
```

```python
df.sort_values(['passing english', 'english'])
```

```python
df['overall grade'] = (df.math + df.english + df.reading) / 3
```

```python
df.sort_values('overall grade')
```

```python
mpg = data('mpg')
```

```python
mpg.shape
#234 rows, 11 columns
```

```python
mpg.info(), mpg.describe()
# Shows all column types
```

```python
mpg = mpg.rename(columns = {'cty': 'city', 'hwy': 'highway'})
mpg
```

```python
mpg[mpg.city > mpg.highway]
# No cars have better city milage to highway mileage
```

```python
mpg['milage difference'] = mpg.highway - mpg.city
mpg
```

```python
mpg.sort_values('milage difference', ascending = False).head()
# Honda civic and Volks New Beetle have highest difference at 12
```

```python
mpg = mpg.rename(columns = {'class': 'type'})
```

```python
mpg[mpg.type == 'compact'].sort_values('milage difference').head(1)
```

```python
mpg[mpg.type == 'compact'].sort_values('milage difference').tail(1)
```

```python
mpg['average milage'] = (mpg.city + mpg.highway) / 2
mpg
```

```python
mpg[mpg.manufacturer == 'dodge'].sort_values('average milage').head(1)
```

```python
mpg[mpg.manufacturer == 'dodge'].sort_values('average milage').tail(1)
```
